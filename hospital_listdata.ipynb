```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import pandas as pd

# Initialize the WebDriver
driver = webdriver.Chrome()

# Open the URL
url = 'https://www.healthfrog.in/hospital/cancer-hospital/rajasthan'
driver.get(url)

# Click the "Load More Records" button a specific number of times
for _ in range(2):  # Change 10 to the desired number of clicks
    try:
        # Wait for the button to be clickable
        load_more_button = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.ID, 'loadmore')))
        load_more_button.click()
        WebDriverWait(driver, 10).until(EC.invisibility_of_element_located((By.ID, 'loadmore')))
    except Exception as e:
        print("Failed to click the button:", e)
        break

# Extracting text from 'a' tags and 'p' tags
data = []
soup = BeautifulSoup(driver.page_source, 'html.parser')
elements = soup.find_all(class_='post p0')
for element in elements:
    links = element.find_all('a')
    paragraphs = element.find_all('p')
    for i in range(min(len(links), len(paragraphs))):
        hospital_name = links[i].text
        address = paragraphs[i].text
        # Extracting mobile number and landline number from the linked page
        if 'href' in str(links[i]):
            link_url = links[i]['href']
            driver.get(link_url)
            try:
                WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, 'fa-phone')))
                soup_inner = BeautifulSoup(driver.page_source, 'html.parser')
                phone_numbers = soup_inner.select('.fa-phone')[0].parent.text.strip()
                data.append((hospital_name, address, phone_numbers))
            except Exception as e:
                print(f"Failed to extract data for {hospital_name} - {address}: {e}")
            finally:
                driver.back()  # Return to the main page

# Creating a DataFrame
df = pd.DataFrame(data, columns=['Hospital Name', 'Address', 'Phone Numbers'])

# Saving the DataFrame to an Excel file
df.to_excel('hospital_cancer_hospital_rajasthan.xlsx', index=False)

print("Data saved successfully.")

# Quit the WebDriver
driver.quit()
```
